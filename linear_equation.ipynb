## Линейна функция

Линейната функция се задава с уравнението  $y=\mathit{ax}+b$ а графиката ѝ, както е известно, е права линия. Това не е трудно да се провери експериментално, като се направят няколко точки върху координатната равнина и се прекара права линия. Такъв експеримент не може да замени строгото математическо доказателство, но пък много добре онагледява
предположението. 

В учебниците такова доказателство обаче липсва, което въобще не трудно да се изведе, само трябва да знаем подобните триъгълници. Така, нека вземем функцията  $y=\mathit{ax}+b$  и заместим две различни стойности в уравнението за $\mathit{x1}$  и  $\mathit{x2}$ , тогава получаваме следните уравнение:

$y_1=\mathit{ax}_1+b$ и <br>
$y_2=\mathit{ax}_2+b$

Ако от първото уравнение извадим второто, получаваме:

\begin{equation*}
y_2-y_1=(\mathit{ax}_1+b)-(\mathit{ax}_2+b)
\end{equation*}
след разкриване на скобите, променливата  $b$ \ се самоунищожава и групираме:

\begin{equation*}
y_1-y_2=a(x_1-x_2)
\end{equation*}
Изразяваме за а и се получава следното:

\begin{equation*}
a=\frac{y_1-y_2}{x_1-x_2}
\end{equation*}
Можем да означим следното  $\Delta x=x_1-x_2$ \ и  $\Delta y=y_1-y_2$, тогава изразяването за ще бъде:

\begin{equation*}
a=\frac{\Delta y}{\Delta y}
\end{equation*}

Нека преди да отидем до доказателството, да разгледаме уравнението и да се опитаме да открием смисълът му. 
<br>
### Физически смисъл

Уравнението  $a=\frac{\Delta y}{\Delta x}$  има следния физически смисъл, ако с променливата $a$, се означи времето, а с променливата  $y$ изминатият път, тогава  $\Delta x=x_1-x_2$  ще обозначава интервала време между два момента ( $x_1$ и  $x_2$), а  $\Delta y=y_1-y_2$  ще бъде изминатият път за посочения интервал от време, тогава уравнението $y=\mathit{ax}+b$ ще описва движението на някакво физическо тяло.


Уравнението  $\Delta y=a{\Delta}x$ означава, че ако вземете еднакви интервали от време, те ще съответстват на едни и същи изминати пътища. Например, ако измервате времето в минути, тялото изминава един и същ път за всяка минута. Ако превърнем минутите в секунди, пътят за всяка секунда ще бъде 60 пъти по-малък от пътя за минута. Всъщност, ако се намали  $\Delta x$  60 пъти, от уравнението   $\Delta y=a{\Delta}x$  следва, че пътят  $\Delta y$  също ще бъде 60 пъти по-малък. Във физиката това движение се нарича равномерно. Отношението  $a=\frac{\Delta y}{\Delta x}$ е отношение на изминатото разстояние към времето, което е необходимо да се измине разстоянието  $\Delta y$ и има значение на скорост във физиката.

<br>

### Геометричен смисъл

За да се открие уравнението на правата линия  $y=\mathit{ax}+b$, през две точки, да речем с координати (1,2) и (5,1),
подхождаме, като заместим в уравнението  $y=\mathit{ax}+b$ вместо x, y със техните координати и отнови получаваме
система от две уравнения:

\begin{equation*}
\left|\begin{matrix}2=a+b\\1=5a+b\end{matrix}\right.
\end{equation*}
решенията на която са, a = -1/4 b= 9/4, и уравнението на правата линия добива вида: $y=\frac{-1} 4x+\frac 9 4$


Има интересен случай обаче, когато системата уравнения няма решение, като например дадени са точки с координати (1,2) и
(1,3). Съставяме отново система, в която са заместени координатите на новите точки, или:

\begin{equation*}
\left|\begin{matrix}2=a+b\\3=a+b\end{matrix}\right.
\end{equation*}
Дори и не е необходимо да се мъчим да решим системата, защото се вижда, че няма как 2 да стане равно на 3. На пръв
поглед е изумително, защото, винаги през две точки може да се прекара една права линия. Оказва се, че уравнението през
точките (1,2) и (1,3) не се записва във вида  $y=\mathit{ax}+b$, а се представя като уравнение x = 1. Това от своя
страна предполага някаква непълнота при записването на уравнението, а именно, че уравнението не е валидно за прави,
които са паралелни на ординатната ос. В този случай, общото уравнение се записа във вида  $c=\mathit{ax}+\mathit{by}$,
което е валидно за всички прави от координатната система.



#### Но да се върнем към доказателството и начертаем един прост чертеж 




Оста  $x$ е разделена на равни части  $\Delta x$, и на всяко  $\Delta x$, съответства една и съща част  $\Delta y$.
Диаграмата представлява стълбица от еднакви триъгълници с катети  ${\Delta}x$ \ и  $\Delta y$, а техните хипотенузи са
построени по протежение на правата. Ако променим хоризонталното разстояние  $\Delta x$, в същото време  $\Delta y$
 също ще се променят със същата стойност, а ъглите на триъгълниците ще останат същите. Следователно хипотенузите им
отново ще бъдат построени по същата права линия. Това означава, че графиката на линейната функция е права линия.
